## Smart Parking Lot Management System

```python
def smart_parking(vehicle_logs,capacity):
    current_peaked=0
    peak_usage=0
    for log in vehicle_logs:
        if log=="IN":
            current_peaked+=1
        elif log=="OUT" and current_peaked>0:
            current_peaked-=1
        if current_peaked>peak_usage:
            peak_usage=current_peaked
            print("currently peaked vehicles:",current_peaked)
            print("peak parking usage:",peak_usage)
            if current_peaked>capacity:
                print("parking status: capacity exceded")
            else:
                print("parking status Available")

vehicle_logs=["IN","IN","IN","OUT","IN","IN","OUT"]
parking_capacity=50
smart_parking(vehicle_logs,parking_capacity)
    
```

## Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    # Base time calculation (5 mins per km)
    eta = distance * 5
    
    # Traffic impact
    if traffic == "High":
        eta += 20
    elif traffic == "Medium":
        eta += 10
    
    # Weather impact
    if weather == "Rainy":
        eta += 15
    elif weather == "Storm":
        eta += 25
    
    print("Estimated Delivery Time:", eta, "minutes")
estimate_delivery_time(8, "High", "Rainy")
```

## Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    
    print("Occupancy:", round(occupancy, 2), "%")
    
    if occupancy == 100:
        print("Show Status: Housefull 🎉")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
total_seats = 200
booked_seats = [1] * 150

seat_occupancy_analyzer(total_seats, booked_seats)
```

## Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    average_cpu = sum(cpu_readings) / len(cpu_readings)
    
    print("Average CPU Load:", round(average_cpu, 2), "%")
    
    if average_cpu < 50:
        print("Server Status: Normal ✅")
    elif 50 <= average_cpu <= 80:
        print("Server Status: Warning ")
    else:
        print("Server Status: Critical ")
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

## Smart Classroom Resource Usage Monitor

```python
def classroom_resource_monitor(resource_usage):
    overused = []
    
    for resource, hours in resource_usage.items():
        if hours > 8:  # Threshold for overuse
            overused.append(resource)
    
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes ")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No ")
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(usage)
```

## Online Event Registration Capacity Controller

```python
def event_registration_controller(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open "
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed "
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
event_registration_controller(100, 105)
```
