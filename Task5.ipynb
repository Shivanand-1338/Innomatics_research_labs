## 1.Social Meadia-Post management Analyzer

```python
def analyze_post_management(likes):
    total_likes=0
    for like in likes:
        total_likes+=like
    if total_likes>=1000:
        status="Viral Post"
    else:
        status="Normal management"
    return total_likes,status
likes=[200,300,400]
total,status=analyze_post_management(likes)
print(f"Total Likes: {total}\n Post status:{status}")
```

## 3.Agriculture Rainfall Adequacy Checker

```python
def check_rainfall(rainfall_data,required_level):
   
    average=sum(rainfall_data)/len(rainfall_data)
    if average>required_level:
        status="Adequate Rainfall"
    else:
        status="InAdequate Rainfall"
    return average,status
rainfall_data=[60,80,75,70]
average,status=check_rainfall(rainfall_data,70)
print("Average rainfall:",average)
print("Rainfall Status: ",status)
```

## 4.Social Media-Duplicate Account Detection

```python
def detect_duplicate_accouts(usernames):
    unique_users=set(usernames)
    if len(unique_users)<len(usernames):
        result="Yes"
    else:
        result="No"
    return result
usernames=["jhon","raj","jhon","meena"]
print("Duplicates Account Found: ",detect_duplicate_accouts(usernames))
```

## 6.Agriculture premium crop filter


```python
def filter_premium_crops(prices):
    premium=[]
    for price in prices:
        if price>2000:
            premium.append(price)
    return premium
prices=[1500,2000,1800,3200]
print("Premium Crops:",filter_premium_crops(prices))
```

## 7. System Monitoring - Application Health checker

```python
def check_application_health(errors):
    if errors==0:
        status="healthy"
    elif errors<=5:
        status="minor issues"
    else:
        status="Critical Issues"
    return status

errors=int(input("Enter the number of errors:"))
print("System status:",check_application_health(errors))
```

## 9.E-Learning Student Eligibility System

```python
def check_attendance(attendance_list):
    total_classes=len(attendance_list)
    attended=0
    for status in attendance_list:
        if status==1:
            attended+=1
    percentage=(attended/total_classes)*100
    if percentage>=75:
        return "Eligible"
    else:
        return "Not eligible"
attendance=[1,1,1,0,1,1,1,1,0,1]
print("Exam eligibility:",check_attendance(attendance))
    
```
