## 1.Unique words in a sentence

```python
def unique_words(sentence):
    words=sentence.split()
    sentence1=set(words)
    return f"unique word count: {len(sentence1)}\nunique words:{sentence1}"
sentence=input("Enter the sentence:")
print(unique_words(sentence))
```

## 2.Highest salary from Employee Data

```python
employees={
    "Ravi":75000,
    "Anita":68000,
    "Kiran":72000
}
highest=max(employees,key=employees.get)
highest_salary=employees[highest]
print(f"Highest Salary: {highest} - {highest_salary}")
```

## 3.Find Maximum and minimum Values


```python
numbers=[45,22,89,10,66]
max=numbers[0]
min=numbers[0]
for num in numbers:
    if num>max:
        max=num
    if num<min:
        min=num
print("List: ",numbers)
print("Max: ",max)
print("Min: ",min)
```

## 4.Count products above a price threshold

```python

def product(prices):
    count=0
    for price in prices:
        if price>1000:
            count+=1
    return f"Products above 1000: {count}"
prices=[450,1200,899,1500,300]
print(product(prices))
```

## 7. Count Character frequency

```python
s="python"
frequency={}
for char in s:
    if char in frequency:
        frequency[char]+=1
    else:
        frequency[char]=1
print(frequency)
```
