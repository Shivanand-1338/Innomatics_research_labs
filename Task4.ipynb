## Mobile Recharge Validation System


```python
def validate_recharge(amount):
    valid_plans=[199,299,399,599]
    if amount<50:
        print("Recharge amount must be >= 50")
        return False
    if amount not in valid_plans:
        print("invalid plan! the valid plans are 199,299,399,599")
        return False
    return True
while True:
    amount=int(input("Enter the amount:"))
    if validate_recharge(amount):
        print("recharge Successful")
        break
    else:
        print("Try again")
```

## Inventory Reorder Alert System

```python
def Reorder(products):
    for product in products:
        if products[product]<15:
            print(f"{product} -> Reorder Alert")
        else:
            print(f"{product} -> stock ok")
    
inventory={
    "rice":14,
    "oil":25,
    "sugar":10,
    "wheat":30
}
Reorder(inventory)
```

## Student Result Processing System

```python
def process_result(marks):
    total=0
    for mark in marks:
        total+=mark
    average_marks=total/len(marks)
    if average_marks>=50:
        print("Pass")
    else:
        print("Fail")
        
marks=[]
n=int(input("Enter the number of subjects:"))
for num in range(n):
    mark=int(input("Enter Mark"))
    marks.append(mark)
print(process_result(marks))
```

## Password Strength Checker

```python
def check_password(password):
    has_digit=False
    has_special=False
    for ch in password:
        if ch.isdigit():
            has_digit=True
        if ch in "@#$":
            has_special=True
    if len(password)>=8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
password=input("Enter the password:")
result=check_password(password)
print("password status:",result)
```

## Employee Attendance Eligibilit Checker

```python
def Eligible(Attendance):
    present_days=0
    for day in Attendance:
        if day=="P":
            present_days=present_days+1
    percentage=(present_days/len(Attendance))*100
    if percentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"
Attendance=[]
n=int(input("Enter the number of days:"))
for i in range(n):
    attend=input("Enter the attendance")
    Attendance.append(attend)
print("Employee Eligibility:",Eligible(Attendance))        
        
```
