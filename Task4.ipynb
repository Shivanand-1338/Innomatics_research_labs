### Mobile Recharge Validation System

```python
# Function to validate recharge amount
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge must be ₹50 or more.")
        return False

    if amount not in valid_plans:
        print("Invalid plan. Valid plans are 199, 299, 399, 599.")
        return False

    return True


# While loop for retry
while True:
    amount = int(input("Enter recharge amount: ₹"))

    if validate_recharge(amount):
        print("Recharge successful!")
        break
    else:
        print("Try again.\n")
```

## Inventory Reorder Alert System

```python
# Function to check stock levels
def check_inventory(products):
    for item in products:
        if products[item] < 15:
            print(item, "- Reorder Alert")
        else:
            print(item, "- Stock OK")


# Dictionary of products and stock quantity
inventory = {
    "Rice": 10,
    "Oil": 25,
    "Sugar": 12,
    "Wheat": 30
}

# Call the function
check_inventory(inventory)
```

# Student Result Processing System


```python
# Function to process marks
def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total = total + mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Taking input from user
marks = []
n = int(input("Enter number of subjects: "))

for i in range(n):
    mark = int(input("Enter mark: "))
    marks.append(mark)

# Get result
result = process_result(marks)

print("Result:", result)
```

# Cab Fare Estimator with Retry Option

```python
# Function to calculate fare
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    # Add 25% extra for peak hour
    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


# Retry using while loop
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)

    print("Total Fare: ₹", total_fare)

    choice = input("Do you want to calculate again? (yes/no): ")
    if choice == "no":
        break

    print()
```

# Employee Attendance Eligibility Checker


```python
# Function to check attendance eligibility
def check_attendance(attendance):
    present_days = 0

    # Count number of "P"
    for day in attendance:
        if day == "P":
            present_days = present_days + 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Taking input from uPA
ser
attendance = []
n = int(input("Enter number of days: "))

for i in range(n):
    status = input("Enter attendance (P/A): ")
    attendance.append(status)

# Get result
result = check_attendance(attendance)

print("Attendance Result:", result)
```

# Password Strength Checker

```python
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False

    # Check each character in password
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    # Check all conditions
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Take input from user
password = input("Enter password: ")

# Get result
result = check_password(password)

print("Password Status:", result)
```
